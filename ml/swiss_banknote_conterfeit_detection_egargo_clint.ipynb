```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
import pandas as pd

# Uncomment if running in Kaggle.
# df = pd.read_csv('/kaggle/input/swiss-banknote-conterfeit-detection/banknotes.csv')

df = pd.read_csv('./datasets/banknotes.csv')
```

```python
X = df.drop(['conterfeit'], axis=1)
y = df[['conterfeit']]
```

```python
X
```

```python
y
```

```python
trainX, testX, trainY, testY = train_test_split(X, y, test_size=0.33)
```

```python
trainX.isnull()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(trainX, trainY)
```

```python
predY = dt.predict(testX)
```

```python
cm = confusion_matrix(testY, predY)
cm
```

```python
tp, fp, fn = cm.ravel()[:3]
tn = cm.ravel()[-1]

precision = tp/(tp+fp)
recall = tp/(tp+fn)
specificity = tn/(fp+tn)
f1= 2* (precision*recall)/(precision+recall)

precision, recall, specificity , f1
```

# SVM

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear')
svm.fit(trainX, trainY.values.ravel())
```

```python
predY = svm.predict(testX)
```

```python
cm = confusion_matrix(testY, predY)
cm
```

```python
tp, fp, fn = cm.ravel()[:3]
tn = cm.ravel()[-1]

precision = tp/(tp+fp)
recall = tp/(tp+fn)
specificity = tn/(fp+tn)
f1= 2* (precision*recall)/(precision+recall)
precision, recall, specificity , f1
```

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(trainX, trainY.values.ravel())
```
